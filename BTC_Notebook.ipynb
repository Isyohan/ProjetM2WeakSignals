```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

import yfinance as yf
from yahoofinancials import YahooFinancials
```

```python
start = dt.datetime(2021, 2, 6)
end = dt.datetime.now()
btc_usd = yf.download('BTC-USD', start=start, end=end, progress=False)
```

```python
btc_usd_grouped = btc_usd.resample('D')['Close'].mean()
btc_usd_grouped

plt.plot(btc_usd_grouped.index, btc_usd_grouped,'-')
plt.show()
```
